### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.

sales_20 = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_21 = pd.read_csv("Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales = pd.concat([sales_20, sales_21], axis = 0)
sales = sales.reset_index(drop=True)

sales
```

```python
# Check if any values are null.
nulls = sales.isnull().any()
nulls
```

```python
# Check the data type of each column
sales.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
"""Use either the groupby or pivot_table function to create a multi-index DataFrame with the "region", "state", 
and "city" columns.

Rename the aggregated column to reflect the aggregation of the data in the column.

Sort the results in ascending order to show the top five regions, including the state and city that have the 
greatest number of products sold. """

## CE - student note - the challenge instructions requests 'ascending' order though it also requests the top five and 
## image provided show a descending list. I modified my table to descend to match the image provided.

# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

total_products = sales.groupby(["region", "state", "city"])[["units_sold"]].agg(["sum"])
                        
#total_products

total_products_flat=total_products.copy()

total_products_flat.columns  = total_products.columns.to_flat_index()

total_products_flat.columns = ['-'.join(column) for column in total_products_flat.columns ]

total_products_flat = total_products_flat.rename(columns={"units_sold-sum":"Total_Products_Sold"}).\
    sort_values("Total_Products_Sold", ascending=False)


# Show the top 5 results.
total_products_flat.head()
                        
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_pivot = pd.pivot_table(sales, 
                                     index = ["region", "state", "city"],
                                     values = "units_sold",
                                     aggfunc = (["sum"]))


level_0 = total_products_pivot.columns.get_level_values(0)
level_1 = total_products_pivot.columns.get_level_values(1)

total_products_pivot.columns = level_0 +"-"+ level_1

#total_products_pivot

# Rename the "units_sold" column to "Total_Products_Sold"

total_products_pivot = total_products_pivot.rename(columns={"sum-units_sold":"Total_Products_Sold"})

total_products_pivot = total_products_pivot.sort_values("Total_Products_Sold", ascending=False)

# # Show the top 5 results.

total_products_pivot.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
"""Use either the groupby or pivot_table function to create a multi-index DataFrame with the "region", "state", and 
"city" columns.

Rename the aggregated column to reflect the aggregation of the data in the column.

Sort the results in ascending order to show the top five regions, including the state and city that generated the most 
sales. 

"""
# Show the total sales for the products sold for each region, state, and city.

total_sales = sales.groupby(["region", "state", "city"])[["total_sales"]].agg(["sum"])


level_0 = total_sales.columns.get_level_values(0)
level_1 = total_sales.columns.get_level_values(1)
total_sales.columns = level_0 +"-"+ level_1

# Rename the "total_sales" column to "Total Sales"

total_sales = total_sales.rename(columns={"total_sales-sum":"Total Sales"})

tot_sales_df = total_sales.sort_values(by="Total Sales", ascending = False)

# Show the top 5 results.
tot_sales_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.

sales_pivot = pd.pivot_table(sales, 
                                 index = ["region", "state", "city"],
                                 values = "total_sales",
                                 aggfunc = (["sum"]))

# Optional: Rename the "total_sales" column to "Total Sales"

level_0 = sales_pivot.columns.get_level_values(0)
level_1 = sales_pivot.columns.get_level_values(1)
sales_pivot.columns = level_0 +"-"+ level_1

sales_pivot = sales_pivot.rename(columns={"total_sales-sum":"Total Sales"})

sales_pivot_df = total_sales.sort_values(by="Total Sales", ascending = False)

# Show the top 5 results.
sales_pivot_df.head()
```
